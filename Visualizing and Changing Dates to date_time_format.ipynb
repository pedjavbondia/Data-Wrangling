# Changing the Date Format of Column 9

```python
#First, I am going to change the format, then I am going ask it to get the weekday and add it to a new column called "day_of_the_week"
```

```python
df['Last_Visited']=pd.to_datetime(df.Last_Visited)
```

```python
df.info()
```

```python
#As you can see the Last_Visited column is changed from object to datetime.
```

```python
df['day_of_the_week']=df.Last_Visited.dt.day_of_week
```

```python
#The code written on line 18 is taking from Last_Visited column the day of the week and placing that info in a new column called "day_of_the_week".
```
